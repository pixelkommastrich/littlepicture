```python
import matplotlib.pyplot as plt
import pandas as pd

# Read CSV file
df = pd.read_csv('../../data/mosaic.csv')

# Extracting data from the DataFrame
years = df['year']
mosaic = df['mosaic']
urban = df['urban']

# Plotting
plt.figure(figsize=(10, 6))

plt.fill_between(years, mosaic, label='Mosaic: > 50% natural cover/< 50% cropland', alpha=0.5)
plt.fill_between(years, urban, label='Urban areas', alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Mosaic Vegetation')
plt.title('european landcover - urban area vs mosaic vegetation 1992 - 2015')
plt.savefig("../../data/mosaic.svg")
# Show plot
plt.show()
```
